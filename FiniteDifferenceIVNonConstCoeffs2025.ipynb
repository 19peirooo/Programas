```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n=100
h=(np.e-1)/(n+1)
indx=np.linspace(0,n+1,n+2)
xvals=1+(indx/(n+1))*(np.e-1)
y0=2
ynplus1=8
```

```python
indx
```

```python
xvals
```

```python
h
```

```python
M=np.identity(n+1)
```

```python
M
```

```python
for i in range(n+1):
    M[i][i]=-2*xvals[i+1]
    
```

Note the indices below.  

```python
for i in range(n):
    M[i+1][i]=xvals[i+2]-h/2
```

```python
for i in range(n):
    M[i][i+1]=xvals[i+1]+h/2
```

```python
M
```

```python
bb=np.zeros(n+1)
```

```python
bb[0]=-(xvals[1]-h/2)*y0
```

```python
bb[n]=-(xvals[n]+h/2)*ynplus1
```

```python
bb
```

```python
y=np.linalg.solve(M,bb)
```

```python
y
```

```python
yb=np.insert(y,0,y0)
```

```python
yb
```

```python
plt.plot(xvals,yb,'g^')
```
