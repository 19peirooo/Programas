```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n=100
h=1/(n+1)
indx=np.linspace(0,n+1,n+2)
xvals=(indx/(n+1))
y0=0
ynplus1=1
```

```python
indx
```

```python
xvals
```

```python
h
```

```python
M=np.identity(n+1)
```

```python
for i in range(n+1):
    M[i][i]=(6*(h**2)-2)
```

```python
for i in range(n):
    M[i+1][i]=(1-h)
```

```python
for i in range(n):
    M[i][i+1]=(1+h)
```

```python
M
```

```python
b=np.zeros(n+1)
```

```python
b[n]=-(1+h)
```

```python
b
```

```python
y=np.linalg.solve(M,b)
```

```python
y
```

Our solution does not include the y0 value so we put it in here.

```python
yb=np.insert(y,0,y0)
```

```python
yb
```

```python
plt.plot(xvals,yb,'g^')
```

Below we compare our numerical solution with the contiuous solution.

```python
cnst=np.e/np.sin(np.sqrt(5))
realanswer=cnst*(np.sin(np.sqrt(5)*xvals))*np.exp(-xvals)
```

```python
plt.plot(xvals,realanswer,'g^',xvals,yb,'r-')
```

Looking at the difference between the real and numerical solutions, you can see how this error changes by increasing n.

```python
plt.plot(xvals,realanswer-yb,'g^')
```
