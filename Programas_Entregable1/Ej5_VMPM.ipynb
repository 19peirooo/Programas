# <b>Solucion</b>

## <i> Condiciones Iniciales </i>
n = 100 <br>
Intervalo: [0,1]

```python
n = 100
h = 1/(n+1)
idx = np.linspace(0,n+1,n+2)
x_vals = idx/(n+1)
y_0 = 0
y_1 = 0
```

Valores de x

```python
x_vals
```

## <i>Montaje de Matriz para Segunda Derivada</i>

$$
y''(x_i) \approx \frac{y_{i+1} -2y_i + y_{i-1}}{h^2}
$$

```python
M = np.identity(n+1)
```

```python
for i in range(n+1):
    M[i][i] = -2
```

```python
for i in range(n):
    M[i+1][i] = 1
    M[i][i+1] = 1
```

```python
M
```

```python
M = M / (h**2)
```

```python
M
```

## <i> Montaje de la Matriz de la Primera Derivada </i>

$$
y'(x_i) \approx \frac{y_{i+1} - y_{i-1}}{2h}
$$

```python
N = np.identity(n+1)
```

```python
for i in range(n+1):
    N[i][i] = 0
```

```python
for i in range(n):
    N[i+1][i] = -1
    N[i][i+1] = 1
```

```python
N
```

```python
N = N / (2*h)
```

```python
N
```

## <i> Resolucion </i>

$$
Ay = \lambda y \rightarrow \lambda y = -(y'' + y')y = -(M+N)y
$$

```python
A = -(M+N)
```

```python
lambdas,vectors = np.linalg.eig(A)
```

```python
lambdas
```

```python
vectors
```

```python
sorted_indices = np.argsort(lambdas)
sorted_eigenvalues = lambdas[sorted_indices]
sorted_eigenvectors = vectors[:,sorted_indices]
```

```python
sorted_eigenvalues
```

```python
yb=np.insert(sorted_eigenvectors[:,1],0,y_0)
```

```python
plt.plot(x_vals,yb,'g^')
```
