```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n=50
h=(2*np.pi)/(n+1)
indx=np.linspace(0,n+1,n+2)
xvals=(indx/(n+1))*2*np.pi
y0=0
ynplus1=0
```

```python
indx
```

```python
xvals
```

```python
h
```

```python
M=np.identity(n+1)
```

```python
for i in range(n+1):
    M[i][i]=-2
```

```python
for i in range(n):
    M[i+1][i]=1
```

```python
for i in range(n):
    M[i][i+1]=1
```

```python
M
```

```python
M=M*(1/h**2)
```

```python
M
```

The differential equations eigenvalue problem now is a regular basic matrix eigenvalue problem.
Luckily we have tools to help us since the matrices are huge.

```python
w,v=np.linalg.eig(M)
```

```python
ew=abs(w)
```

```python
ew
```

```python
v
```

```python
sorted_indices = np.argsort(ew)
sorted_eigenvalues = ew[sorted_indices]
sorted_eigenvectors = v[:,sorted_indices]
```

```python
sorted_eigenvalues
```

```python
yb=np.insert(sorted_eigenvectors[:,1],0,y0)
```

```python
plt.plot(xvals,yb,'g^')
```

```python
sorted_eigenvalues[1]
```
