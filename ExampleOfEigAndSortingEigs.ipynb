```python
import numpy as np


mat = np.mat("1 2 3;1 3 4;3 2 1")

# Original matrix
print(mat)
print("")
evalue, evect = np.linalg.eig(mat)

# Eigenvalues of the said matrix"
print(evalue)
print("")

# Eigenvectors of the said matrix
print(evect)

# NOTE that the eigenvectors are the COLUMNS of the evect matrix.
# In the example below the vector (-0.42932,0.7945835,-0.42932) is the eigenvector corresponding
# to the eigenvalue 0.29843788
```

```python
evect[:,0]
```

```python
sorted_indices = np.argsort(evalue)
sorted_eigenvalues = evalue[sorted_indices]
sorted_eigenvectors = evect[:,sorted_indices]
```
